```python
import pandas as pd

alb = pd.read_csv("../ERA5Albanyhourly2mTemp.csv")
chi = pd.read_csv("../ERA5Chicagohourly2mTemp.csv")

print("Albany columns:", alb.columns.tolist())
print("Chicago columns:", chi.columns.tolist())

alb_time = "valid_time"
chi_time = "valid_time"

alb[alb_time] = pd.to_datetime(alb[alb_time])
chi[chi_time] = pd.to_datetime(chi[chi_time])

alb_sample = alb[(alb[alb_time] >= "2020-01-01") & (alb[alb_time] < "2020-03-01")]
chi_sample = chi[(chi[chi_time] >= "2020-01-01") & (chi[chi_time] < "2020-03-01")]

alb_sample.to_csv("Albany_temp_sample.csv", index=False)
chi_sample.to_csv("Chicago_temp_sample.csv", index=False)

mjo = pd.read_csv(
    "../rmm.74toRealtime.txt",
    sep=r"\s+",
    skiprows=2,
    header=None,
    names=["year", "month", "day", "rmm1", "rmm2", "phase", "amplitude", "source"]
)

mjo["date"] = pd.to_datetime(mjo[["year", "month", "day"]])

mjo_sample = mjo[(mjo["date"] >= "2020-01-01") & (mjo["date"] < "2020-03-01")]
mjo_sample.to_csv("mjo_sample.csv", index=False)
```
